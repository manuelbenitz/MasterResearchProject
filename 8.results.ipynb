## **4. MERGE INTO DATAFRAMES PER EVALUATION METRIC**

```python
# Process Target column
def preprocess_t(dataframe):
    if dataframe.index.name == "Target":
        dataframe = dataframe.reset_index()
    dataframe = dataframe.loc[:, ~dataframe.columns.duplicated()]
    return dataframe

baseline = preprocess_t(baseline)
lstm_coords = preprocess_t(lstm_coords)
lstm_nocoords = preprocess_t(lstm_nocoords)
cnnlstm_coords = preprocess_t(cnnlstm_coords)
cnnlstm_nocoords = preprocess_t(cnnlstm_nocoords)

metrics = ["MAE", "MAPE", "RMSE", "R2"]

# Filter columns by metric
def fbymetric(dataframe, metric):
    cols = ["Target"] + [col for col in dataframe.columns if metric in col]
    return dataframe[cols]

# Dictionary with all results
results = {}

for metric in metrics:
    merged = (
        fbymetric(baseline, metric)
        .merge(fbymetric(lstm_nocoords, metric), on="Target", how="inner", suffixes=("_baseline", "_lstm_nocrds"))
        .merge(fbymetric(lstm_coords, metric), on="Target", how="inner", suffixes=("", "_lstm_coords"))
        .merge(fbymetric(cnnlstm_nocoords, metric), on="Target", how="inner", suffixes=("", "_cnnlstm_nocrds"))
        .merge(fbymetric(cnnlstm_coords, metric), on="Target", how="inner", suffixes=("", "_cnnlstm_coords"))
    )
    results[metric] = merged
```

# **4. PREDICTIONS**

## **4.1 BASELINE (RANDOM FOREST AND XGBOOST)**

```python
# Establish connection to counters dataset
CONTAINER_NAME = "predictions"
container_client = blob_service_client.get_container_client(CONTAINER_NAME)

# Random Forest
file_name = "pred_rf.csv"
pred_rf = load_csv_from_blob(file_name)
print(f"Loaded '{file_name}' with shape: {pred_rf.shape}")

# XGboost
file_name = "pred_xgb.csv"
pred_xgb = load_csv_from_blob(file_name)
print(f"Loaded '{file_name}' with shape: {pred_xgb.shape}")
```

## **4.2 DEEP LEARNING MODELS (LSTM & CNN-LSTM)**

```python
# Establish connection to counters dataset
CONTAINER_NAME = "predictions"
container_client = blob_service_client.get_container_client(CONTAINER_NAME)

# LSTM no coordinates
file_name = "pred_lstm_nocoords.csv"
pred_lstm_nocoords = load_csv_from_blob(file_name)
print(f"Loaded '{file_name}' with shape: {pred_lstm_nocoords.shape}")

# LSTM with coordinates
file_name = "pred_lstm_coords.csv"
pred_lstm_coords = load_csv_from_blob(file_name)
print(f"Loaded '{file_name}' with shape: {pred_lstm_coords.shape}")

# CNN-LSTM no coordinates
file_name = "pred_cnnlstm_nocoords.csv"
pred_cnnlstm_nocoords = load_csv_from_blob(file_name)
print(f"Loaded '{file_name}' with shape: {pred_cnnlstm_nocoords.shape}")

# CNN-LSTM with coordinates
file_name = "pred_cnnlstm_coords.csv"
pred_cnnlstm_coords = load_csv_from_blob(file_name)
print(f"Loaded '{file_name}' with shape: {pred_cnnlstm_coords.shape}")
```

## **4.3 MERGING ALL PREDICTIONS INTO A COMMON DATAFRAME**

```python
# Consistency in datatypes
for dataframe in [pred_rf, pred_xgb, pred_lstm_nocoords, pred_lstm_coords, pred_cnnlstm_nocoords, pred_cnnlstm_coords]:
    dataframe["date and time"] = pd.to_datetime(dataframe["date and time"], errors="coerce")

# Merge all data using date and time and Target as join
allpreds = (
    pred_rf
    .merge(pred_xgb, on=["date and time", "Target", "Real"], how="inner")
    .merge(pred_lstm_nocoords, on=["date and time", "Target", "Real"], how="inner")
    .merge(pred_lstm_coords, on=["date and time", "Target", "Real"], how="inner")
    .merge(pred_cnnlstm_nocoords, on=["date and time", "Target", "Real"], how="inner")
    .merge(pred_cnnlstm_coords, on=["date and time", "Target", "Real"], how="inner")
)
```

# **DATAFRAMES TO AZURE**

```python
def save_blob(data, filename):
    try:
        blob_name = f"{CONTAINER_NAME}/{filename}"
        csv_data = data.to_csv(index=False)
        with fs.open(blob_name, "w") as f:
            f.write(csv_data)
        print(f"Saved to {blob_name}")
    except Exception as e:
        print(f"Error saving data to blob storage: {str(e)}")
        return False
```

```python
CONTAINER_NAME = "visualisation"

save_blob(mae, "m_mae.csv")
save_blob(mape, "m_mape.csv")
save_blob(r2, "m_r2.csv")
save_blob(rmse, "m_rmse.csv")
```

```python
CONTAINER_NAME = "visualisation"

save_blob(allpreds, "all_preds.csv")
```
